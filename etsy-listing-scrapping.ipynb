# Load Libraries

```python
import pandas as pd

# For text processing
!pip install KeyBERT
from keybert import KeyBERT
import spacy
from collections import Counter
nlp = spacy.load("en_core_web_sm")

# For scrapping
from bs4 import BeautifulSoup
import requests
```

# Load Data

```python
# List down links
urls = ['https://www.etsy.com/c/home-and-living/home-decor/candles-and-holders/candles/container-candles?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=candles&ref=breadcrumb_listing&frs=1&pop=1&sts=1&organic_search_click=1&explicit=1'
, 'https://www.etsy.com/c/home-and-living/home-decor/candles-and-holders/candles/container-candles?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=candles&ref=breadcrumb_listing&frs=1&pop=1&sts=1&organic_search_click=1&explicit=2' 
, 'https://www.etsy.com/c/home-and-living/home-decor/candles-and-holders/candles/container-candles?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=candles&ref=breadcrumb_listing&frs=1&pop=1&sts=1&organic_search_click=1&explicit=3'
, 'https://www.etsy.com/c/home-and-living/home-decor/candles-and-holders/candles/container-candles?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=candles&ref=breadcrumb_listing&frs=1&pop=1&sts=1&organic_search_click=1&explicit=4'
, 'https://www.etsy.com/c/home-and-living/home-decor/candles-and-holders/candles/container-candles?ga_order=most_relevant&ga_search_type=all&ga_view_type=gallery&ga_search_query=candles&ref=breadcrumb_listing&frs=1&pop=1&sts=1&organic_search_click=1&explicit=5']

# Create a table to store the data
df = pd.DataFrame(columns = ['Link','Name','Price','Store','Reviews'])

# Loop across link in the url list
for url in urls: 
  # Scrap a page
  response = requests.get(url)
  soup=BeautifulSoup(response.content,'lxml')

  # Look for specific value in the identified tab
  for i in soup.select('.v2-listing-card__info'):
    link = soup.find('a',{'class':'listing-link'}, href = True)['href']
    listingName = i.select('h3')[0].text.strip()
    listingPrice = i.select('.currency-value')[0].text.strip()
    storeName = i.find('p',{'aria-role':'paragraph'}).text.strip()
    reviews = i.find('span',{'class':' wt-text-body-01 wt-text-gray wt-display-inline-block '})
    numReview = reviews.text.strip() if reviews else "None"
    
    # Add the data into the table
    df.loc[len(df.index)] = [link, listingName, listingPrice, storeName, numReview]

# Print the table
df 
```

# Extract Words
## Word Frequency

```python
# Create lists of product name
docs = list(nlp.pipe(df.Name))

# Extract top 10 highest frequency Noun
nouns = []
for doc in docs:
  nouns+=[token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct and token.is_alpha and token.pos_ == "NOUN"]
print(Counter(nouns).most_common(10))

# Extract top 10 highest frequency Proper Noun
proper_nouns = []
for doc in docs:
  proper_nouns+=[token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct and token.is_alpha and token.pos_ == "PROPN"]
print(Counter(proper_nouns).most_common(10))

# Extract top 10 highest frequency Adjective
adjectives = []
for doc in docs:
  adjectives+=[token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct and token.is_alpha and token.pos_ == "ADJ"]
print(Counter(adjectives).most_common(10))
```

## Words Extraction

```python
kw_model = KeyBERT()
doc = ' '.join(df['Name'].tolist())
kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 2), stop_words='english', diversity=0.5, use_mmr=True, top_n = 20)
```
